```python
# Q1. Find the number of matches played in Mumbai in the period of 2013 and 2017
df[((df['season']>2012) & (df['season']<2018)) & (df['city']=="Mumbai")].count()['id']
```

```python
# Q3. What percent times KKR decides to bat after winning the toss
((df[((df['team1']=="Kolkata Knight Riders") | (df['team2']=="Kolkata Knight Riders")) & (df['toss_decision']=="bat")].count()['id'])*100)/df[(df['team1']=="Kolkata Knight Riders") | (df['team2']=="Kolkata Knight Riders")].count()['id']
```

```python
# Q5 : Find the team who has won most number of matches with victory margin > 50 [Easy]
df2=df[df['win_by_runs']>50]
```

```python
# Q6 : Write a function which will take 2 inputs, team 1 and team 2 and their record against each other
def record(team1,team2):
    df3=df[((df['team1']==team1) & (df['team2']==team2))|((df['team1']==team2) & (df['team2']==team1))]
    total=df3.count()['id']
    t1win=df3[df3['winner']==team1].count()['id']
    t2win=df3[df3['winner']==team2].count()['id']
    draw=total-(t1win+t2win)
    print("total matches :",total)
    print("{} wins :{}".format(team1,t1win))
    print("{} wins :{}".format(team2,t2win))
    print("match drawn :",draw)
```

```python
record("Kolkata Knight Riders","Royal Challengers Bangalore")
```

```python
# Q7 : Write a function which will accept a team name as input and will return it's win percentage after winning the toss[Difficult] 
def winp(team):
    total=df[df['toss_winner']==team].count()['id']
    optimalcase=df[(df['toss_winner']==team) & (df['winner']==team)].count()['id']
    print("win percentage after winning the toss :",round((optimalcase*100)/total))
```

```python
winp("Kolkata Knight Riders")
```
